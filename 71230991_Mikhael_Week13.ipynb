```python
n = int(input("Masukkan jumlah kategori: "))

data = {}

for i in range(n):
    kategoriApp = input(f"Masukkan nama kategori ke-{i+1}: ")
    aplikasi = []
    print(f"Masukkan 5 nama aplikasi untuk kategori '{kategoriApp}': ")

    for j in range(5):
        namaAplikasi = input(f"Aplikasi ke-{j+1}: ")
        aplikasi.append(namaAplikasi)
    data[kategoriApp] = aplikasi
    
print("\nDaftar nama aplikasi di setiap kategori: ")
for kategori, aplikasi in data.items():
    print(f"{kategori}: {", ".join(aplikasi)}")

daftar_aplikasi_list = []
for aplikasi in data.values():
    daftar_aplikasi_list.append(set(aplikasi))

hasil = daftar_aplikasi_list[0]
for i in range(1, len(daftar_aplikasi_list)):
    hasil = hasil ^daftar_aplikasi_list[i]

apk_list = daftar_aplikasi_list[0]
for i in range(1,len(daftar_aplikasi_list)):
    apk_list.update(apk_list & daftar_aplikasi_list[i])
for i in apk_list: 
    if i in hasil:
        hasil.remove(i)
print(hasil)

print("Aplikasi yang hanya muncul di satu kategori saja: ", hasil)

aplikasi_listo = list()
dua_kategori=set()

if n > 2:
    for i in data.values():
        aplikasi_listo += i
    for i in aplikasi_listo:
        if aplikasi_listo.count(i) == 2:
            dua_kategori.add(i)
    print("Aplikasi yang muncul tepat di dua kategori: ", dua_kategori)
```

```python
def ubahFormat():
    listo = ["Makan", "Minum", "tidur", "makan", "belajar"]
    print(f"list : {listo}")
    jadiSet = set(listo)
    print(f"list menjadi set: {jadiSet}")
    
    print(f"\nset : {jadiSet}")
    listo2 = list(jadiSet)
    print(f"set menjadi list : {listo2}")

    tupelbrok = ["Makan", "Minum", "tidur", "tidur", "belajar"]
    print(f"\ntuple : {tupelbrok}")
    jadiSet2 = set(tupelbrok)
    print(f"tuple menjadi set: {jadiSet2}")
    
    print(f"\nset : {jadiSet2}")
    t2 = tuple(jadiSet2)
    print(f"tuple menjadi tuple : {t2}")

ubahFormat()
```
